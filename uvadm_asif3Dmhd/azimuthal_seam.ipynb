```python
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x, y, z, vx, vy, vz, bx, by, bz, rho, T = np.genfromtxt('cube240_T1oc_1Ms.dat', skip_header=17, unpack='true')
#x, y, z, vx, vy, vz, bx, by, bz, rho, T = np.genfromtxt('cube200_T1oc_1Ms.dat', skip_header=17, unpack='true')
#x, y, z, vx, vy, vz, bx, by, bz, rho = np.genfromtxt('for_christi_cube100.dat', skip_header=16, unpack='true')
print(x.shape)
```

```python
rsun = 6.957e10 #cm
rstar = 8*rsun #cm
vinf = 3200 # km/s
bpole = 1100 #G
mdot = 3.3e-7 # Msun/yr

rho_wind = mdot * 2e33 / 3e7 / (4*np.pi*vinf*1e5*rstar**2) # In g/cm3
#(print(rho_wind))

nx = 240
ny = 240
nz = 240

x = np.reshape(x, (nx,ny,nz), order='F')/rstar
y = np.reshape(y, (nx,ny,nz), order='F')/rstar
z = np.reshape(z, (nx,ny,nz), order='F')/rstar
vx = np.reshape(vx, (nx,ny,nz), order='F')/vinf/1e5
vy = np.reshape(vy, (nx,ny,nz), order='F')/vinf/1e5
vz = np.reshape(vz, (nx,ny,nz), order='F')/vinf/1e5
bx = np.reshape(bx, (nx,ny,nz), order='F')/bpole
by = np.reshape(by, (nx,ny,nz), order='F')/bpole
bz = np.reshape(bz, (nx,ny,nz), order='F')/bpole
rho = np.reshape(rho, (nx,ny,nz), order='F')/rho_wind
```

```python
#np.save('numpy_files/x_coord_cube200', x)
#np.save('numpy_files/y_coord_cube200', y)
#np.save('numpy_files/z_coord_cube200', z)

np.save('numpy_files/x_coord_cube240', x)
np.save('numpy_files/y_coord_cube240', y)
np.save('numpy_files/z_coord_cube240', z)
```
