```python
import numpy as np

from matplotlib import pyplot as plt
%matplotlib inline
```

```python
#s1: dxp1, dyp1, dzp1, lp1
wave_less = np.genfromtxt('adm_code_v5/resolution_testing/a90_dxp1_dyp1_dzp1_lp1/wavelengths_vvinf.dat', skip_header=1, unpack=True)
s1_abs = np.genfromtxt('adm_code_v5/resolution_testing/a90_dxp1_dyp1_dzp1_lp1/prof_abs.dat', skip_header=1, unpack=True)
s1_emit = np.genfromtxt('adm_code_v5/resolution_testing/a90_dxp1_dyp1_dzp1_lp1/prof_emit.dat', skip_header=1, unpack=True)

#s2: dxp05, dyp05, dzp05, lp05
wave = np.genfromtxt('adm_code_v5/resolution_testing/a90_dxp05_dyp05_dzp05_lp05/wavelengths_vvinf.dat', skip_header=1, unpack=True)
s2_abs = np.genfromtxt('adm_code_v5/resolution_testing/a90_dxp05_dyp05_dzp05_lp05/prof_abs.dat', skip_header=1, unpack=True)
s2_emit = np.genfromtxt('adm_code_v5/resolution_testing/a90_dxp05_dyp05_dzp05_lp05/prof_emit.dat', skip_header=1, unpack=True)

#s3: dxp05, dyp05, dzp01, lp05
s3_abs = np.genfromtxt('adm_code_v5/resolution_testing/a90_dxp05_dyp05_dzp01_lp05/prof_abs.dat', skip_header=1, unpack=True)
s3_emit = np.genfromtxt('adm_code_v5/resolution_testing/a90_dxp05_dyp05_dzp01_lp05/prof_emit.dat', skip_header=1, unpack=True)

#s4: dxp05, dyp05, dzp05, lp01
wave_more = np.genfromtxt('adm_code_v5/resolution_testing/a90_dxp05_dyp05_dzp05_lp01/wavelengths_vvinf.dat', skip_header=1, unpack=True)
s4_abs = np.genfromtxt('adm_code_v5/resolution_testing/a90_dxp05_dyp05_dzp05_lp01/prof_abs.dat', skip_header=1, unpack=True)
s4_emit = np.genfromtxt('adm_code_v5/resolution_testing/a90_dxp05_dyp05_dzp05_lp01/prof_emit.dat', skip_header=1, unpack=True)

#s5: dxp05, dyp05, dzp01, lp01
s5_abs = np.genfromtxt('adm_code_v5/resolution_testing/a90_dxp05_dyp05_dzp01_lp01/prof_abs.dat', skip_header=1, unpack=True)
s5_emit = np.genfromtxt('adm_code_v5/resolution_testing/a90_dxp05_dyp05_dzp01_lp01/prof_emit.dat', skip_header=1, unpack=True)
```

```python
fig = plt.figure(figsize=(7,7))
plt.rcParams.update({'font.size': 14,'font.family':'Times new roman'})

plt.xlim(-2.0,1.5)
#plt.xlabel('$\lambda \; (v/v_{\infty})$')
plt.ylim(0.0,1.4)
#plt.ylabel('$F_{\lambda}/F_{C}$')

# Plot Line Center
plt.axvline(x=0, ls='--', color='gray')

#Plots 

plt.plot(-wave_less,s1_abs+s1_emit, color='gray', ls='-', label='p1')#, marker='o')
plt.plot(-wave,s2_abs+s2_emit, color='coral', ls='-', label='p05')#, marker='o')
#plt.plot(-wave,s3_abs+s3_emit, color='orange', ls='-', label='dzp01, lp05')#, marker='o')
#plt.plot(-wave_more,s4_abs+s4_emit, color='cornflowerblue', ls='-', label='dzp05, lp01')#, marker='o')
#plt.plot(-wave_more,s5_abs+s5_emit, color='purple', ls='-', label='dzp01, lp01')#, marker='o')


plt.title('a=90.0, k=1.0, x=0.01, Ra=2.7, O+C Loops')
plt.legend(loc='best')

# End Matter
plt.tight_layout()
plt.show()
```
